```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
print(df.shape)   
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

```python
df.dropna(subset=['Postal Code'], inplace=True)
```

```python

df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date']  = pd.to_datetime(df['Ship Date'], dayfirst=True)
```

```python
df['Sales'] = df['Sales'].astype(float)
```

```python
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
print("Duplicates:", df.duplicated().sum())

# Drop exact duplicates
df.drop_duplicates(inplace=True)

print(df['Order ID'].value_counts().head())
```

```python
text_cols = ['Ship Mode', 'Segment', 'Country', 'City', 
             'State', 'Region', 'Category', 'Sub-Category']

for col in text_cols:
    df[col] = df[col].str.strip().str.title()

# Verify unique values for key categories
print(df['Segment'].unique())
print(df['Ship Mode'].unique())
print(df['Category'].unique())
```

```python
# Check Sales distribution
print(df['Sales'].describe())

# Visualize
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=df['Sales'])
plt.title("Sales Outliers")
plt.show()

# IQR Method
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['Sales'] < Q1 - 1.5*IQR) | (df['Sales'] > Q3 + 1.5*IQR)]
print(f"Outliers found: {len(outliers)}")
# Don't drop — sales outliers are often real big orders
```

```python
# Shipping duration
df['Ship Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

# Time features
df['Order Year']    = df['Order Date'].dt.year
df['Order Month']   = df['Order Date'].dt.month
df['Order Quarter'] = df['Order Date'].dt.quarter
```

```python
print("Shape:", df.shape)
print("Nulls:\n", df.isnull().sum())
print("Dtypes:\n", df.dtypes)
print(df.head())

# Save cleaned data
df.to_csv("sales_cleaned.csv", index=False)
```
